Cleaning data 
Cleaning data in pandas means fixing or removing incorrect, missing, or inconsistent data to make the dataset ready for analysis

```python
import pandas as pd
```

```python
# ============================== Pandas Data Cleaning & Preprocessing ==============================

# --- Missing Data ---
# | Operation                         | Description                                 |
# |----------------------------------|---------------------------------------------|
# | df.isnull()                      | Check for missing values                    |
# | df.notnull()                     | Check for non-missing values                |
# | df.dropna()                      | Drop rows with missing values               |
# | df.dropna(axis=1)                | Drop columns with missing values            |
# | df.fillna(value)                 | Fill missing values with a specified value  |
# | df.fillna(method='ffill')       | Forward fill (last valid value)             |
# | df.fillna(method='bfill')       | Backward fill (next valid value)            |
# | df.interpolate()                | Interpolate missing values                  |

# --- Duplicate Data ---
# | df.duplicated()                 | Check for duplicate rows                    |
# | df.drop_duplicates()           | Drop duplicate rows                         |

# --- Value Replacement ---
# | df.replace('old', 'new')        | Replace a specific value                    |
# | df.replace([1, 2], [3, 4])      | Replace multiple values                     |
# | df.replace({1: 'one'})          | Replace using a dictionary                  |

# --- Renaming & Reindexing ---
# | df.rename(columns={"old": "new"}) | Rename columns                            |
# | df.set_index('col')             | Set a specific column as index             |
# | df.reset_index(drop=True)       | Reset index to default                     |

# --- String Cleanup ---
# | df['col'].str.strip()           | Remove leading/trailing whitespace          |
# | df['col'].str.lower()           | Convert strings to lowercase                |
# | df['col'].str.upper()           | Convert strings to uppercase                |
# | df['col'].str.replace(',', '')  | Remove commas or specific characters        |
# | df['col'].str.contains('txt')   | Filter rows containing a pattern            |

# --- Type Conversion ---
# | df.astype({'col': 'int'})       | Convert to specific type                    |
# | pd.to_numeric(df['col'])        | Convert string numbers to numeric           |
# | pd.to_datetime(df['date_col'])  | Convert string to datetime                  |

# --- Filtering & Removing ---
# | df[df['col'] != 'val']          | Remove rows with specific value             |
# | df.query('col > 100')           | Filter rows using a condition               |
# | df[df['col'].isin(['A','B'])]   | Keep rows with values in list               |

# --- Column Cleanup ---
# | df.drop(['col1'], axis=1)       | Drop column(s)                              |
# | df.columns.str.strip()          | Strip whitespace from column names          |
#---------------------------------------------------------------------------------|


# Axis Used: 
# axis=0 ðŸ‘‰ Refers to rows
# axis=1 ðŸ‘‰ Refers to columns
```

```python
# Drop columns with any missing values
exceldata.dropna(axis=1)
```

```python
# Fill missing values with a fixed value
exceldata.fillna(0)
```

```python
# Fill missing values forward (previous row value)
exceldata.fillna(method='ffill')
```

```python
# Fill missing values backward (next row value)
exceldata.fillna(method='bfill')
```

```python
# Interpolate missing values
exceldata.interpolate()
```

```python
# Check for duplicate rows
exceldata.duplicated()
```

```python
# Check for total duplicate rows
exceldata.duplicated().sum()
```

```python
# Replace specific value
exceldata.replace('Lower Manhattan', 'Downtown NYC')
```

```python
# Replace multiple values using a list
exceldata.replace([3.00, 3.10], [3.5, 3.6])
```

```python
# Replace using dictionary
exceldata.replace({'Gourmet brewed coffee': 'Premium Coffee'})
```

```python
# Rename column
exceldata.rename(columns={"product_detail": "product_description"})
```

```python
# Set index to a specific column
exceldata.set_index('transaction_id')
```

```python
# Reset index to default integers
exceldata.reset_index(drop=True)
```
