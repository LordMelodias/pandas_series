```python
import pandas as pd

# Load the Coffee.xlsx data
df = pd.read_excel('Coffee.xlsx')
```

```python
df.head()
```

```python
# Basic line plot
df.plot(kind='line', x='transaction_date', y='transaction_qty', title="Transaction Quantity Over Time")
```

```python
# Bar plot
df.head(10).plot(kind='bar', x='product_category', y='unit_price', title="Transaction Quantity by Product Category")
```

```python
# Histogram
df['transaction_qty'].plot(kind='hist', bins=10, title="Histogram of Transaction Quantities")
```

```python
ax = df.plot(kind='line', x='transaction_date', y='transaction_qty')
ax.set_xlabel('Transaction Date')
ax.set_ylabel('Transaction Quantity')
ax.set_title('Transaction Quantity Over Time')
ax.legend(['Transaction Quantity'])
ax.grid(True)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df.boxplot(column="unit_price")

plt.title("Boxplot of Unit Price")
plt.show()
```

```python
# plt.figure(figsize=(10,10))  # Make the pie chart larger
df["store_location"].value_counts().plot(kind="pie", autopct="%1.1f%%", title="Store Location")
plt.ylabel("")  # Hide y-label
plt.show()
```

```python
# Line plot using Seaborn
sns.lineplot(x='transaction_date', y='transaction_qty', data=df)
```

```python
# Bar plot using Seaborn
sns.barplot(x='product_category', y='transaction_qty', data=df)
```

```python
# Scatter plot using Seaborn
sns.scatterplot(x='transaction_qty', y='unit_price', data=df)
```

```python
# If you want to create separate plots for different subsets based on a specific column
sns.FacetGrid(df, col='product_category').map(plt.scatter, 'transaction_qty', 'unit_price')
plt.show()
```

```python
# Basic Matplotlib Line Plot
plt.plot(df['transaction_date'], df['transaction_qty'], color='orange')
plt.title('Transaction Quantity Over Time')
plt.xlabel('Transaction Date')
plt.ylabel('Transaction Quantity')
plt.grid(True)
plt.show()
```

```python
# Matplotlib Histogram
plt.hist(df['transaction_qty'], bins=20, color='blue')
plt.title('Transaction Quantity Histogram')
plt.xlabel('Transaction Quantity')
plt.ylabel('Frequency')
plt.show()
```
